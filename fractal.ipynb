Импортируем необходимые библиотеки

```python
import math
import numpy
import matplotlib
import matplotlib.pyplot
```

Введём функцию, осуществляющую приближённый поиск корня уравнения x^3 = 1 путём десяти итераций по методу Ньютона

```python
def f(x):
    for i in range (10):
        x -= (x ** 3 - 1) / (3 * x ** 2)
    return x
```

Введём функцию, показывающую, какие из каких точек комплексной плоскости мы придём к решению 1 (красный цвет), из каких - к решению 0.5 + 0.866i (зелёный), а из каких к решению 0.5 - 0.866i (желтый)

```python
def graph(r, eps):
    x = numpy.arange(-r, r, eps)
    y = numpy.arange(-r, r, eps)
    x1 = []
    y1 = []
    x2 = []
    y2 = []
    x3 = []
    y3 = []
    for xi in x:
        for yi in y:
            k = f(xi + yi * 1j).imag
            if k <= -0.8:
                x1 += [xi]
                y1 += [yi]
            elif k >= 0.8:
                x2 += [xi]
                y2 += [yi]
            else:
                x3 += [xi]
                y3 += [yi]
                
    figure, regular_plot = matplotlib.pyplot.subplots()
    regular_plot.plot(x1, y1, 'g,')
    regular_plot.plot(x2, y2, 'y,')
    regular_plot.plot(x3, y3, 'r,')
    matplotlib.pyplot.show()
```

Построим несколько фракталов на разных масштабах и с разной плотностью точек

```python
graph(1, 0.01)
```

```python
graph(0.1, 0.0001)
```

```python
graph(10, 0.1)
```
