```python
import numpy
from matplotlib import pyplot
import math
%matplotlib inline

#Наша функция
def f(x):
    return 1.0 / (1.0 + 25 * x ** 2)

#Лагранжева интерполяция
def lagrange (xval, xk, yk):
    s = 0
    for k in range (len(xk)):
        s += yk[k] * lagrange_k(xval, xk, k)
    return s

def lagrange_k(xval, xk, k):
    n = len(xk)
    x1, x2 = 1.0, 1.0 
    for j in range(n):
        if j == k: 
            continue
        x2 *= (xk[k] - xk[j])
        x1 *= (xval - xk[j])
    return x1 / x2

def plot_all(n):
    x = numpy.arange(-1, 1, 0.01)
    
    #Равномерная сетка
    ravnx = numpy.linspace(-1, 1, n)
    ravny = [f(x) for x in ravnx]
    
    #Оптимальная сетка
    chebx = [math.cos((2 * x + 1) / 2 / n * math.pi) for x in range(n)]
    cheby = [f(x) for x in chebx]
    
    pyplot.plot(x, lagrange(x, ravnx, ravny), 'g')
    pyplot.plot(x, lagrange(x, chebx, cheby), 'b')
    pyplot.plot(x, f(x), 'r')
    """Зеленый график - равномерная сетка, синий - оптимальная, красный - сама функция"""
```

```python
plot_all(3)
```

```python
plot_all(7)
```

```python
plot_all(10)
```

```python
plot_all(13)
```

```python
plot_all(18)
```

При небольшом числе точек отклонения методов интерполяции жруг от друга малы
С увеличением числа точек колебания интерполяции по равномерной сетке начинают значительно отклоняться от исходного графика, а по оптимальной - приближаться к нему
