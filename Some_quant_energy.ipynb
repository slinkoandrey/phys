```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

a = 1
n = 1000
H = np.zeros([n,n])
for j in range(n-2):
    i = j + 1
    H[i,i] = -2
    H[i, i - 1] = H[i, i + 1] = 1
H = H / (a / n)**2
    
k = -np.linalg.eig(H)[0]
k = np.sort(k)
#k = np.sqrt(k)

true = [(np.pi * i)**2  for i in range(len(k))]

plt.xscale('log')
plt.yscale('log')
plt.grid()
plt.plot(k, color = 'red')
plt.plot(true, color = 'blue')
plt.show()
```
